```python
# import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import IsolationForest
```

```python
# load the dataset
df= pd.read_csv('embedded_system_network_security_dataset.csv')
df.head()
```

```python
df.info()
```

```python
# check for missing values
df.isnull().sum()
```

### Step 2: Data Preprocessing

```python
# drop the label column if exists
features = df.drop(columns=['label'])
features.head()
```

```python
# if columns is bool type then convert it to int type
for col in df.columns:
    if df[col].dtype == 'bool':
        df[col] = df[col].astype(int)
```

```python
df.dtypes
```

```python
features.columns
```

### Feature Engineering

```python
# scale the features
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
scaled_features=scaler.fit_transform(features)
```

```python
# convert the scaled features to a dataframe
scaled_features_df=pd.DataFrame(scaled_features, columns=features.columns)
scaled_features_df.head()
```

```python
# parameters for isolation forest
n_estimators=100 # number of trees in the forest
contamination=0.1 # proportion of outliers in the data
sample_size=256 # number of samples to draw from the data to train each tree
```

```python
model = IsolationForest(n_estimators=n_estimators, contamination=contamination,
max_samples=sample_size, random_state=42)

model.fit(scaled_features_df)

# predict the anomalies
anomaly_labels = model.predict(scaled_features_df)

# convert the anomaly labels to binary (1 for normal, -1 for anomaly)
binary_labels = np.where(anomaly_labels == 1, 0,1)
# add the binary labels to the original dataframe
df['anomaly'] = binary_labels
```
